```python
import os
import matplotlib.pyplot as plt
import numpy as np

import matplotlib

font = {'family': 'Arial', 'weight': 'normal', 'size': 14}
matplotlib.rc('font', **font)
LineWidth = 3

bpp = [0.013171296, 0.008599537, 0.016753472, 0.042980324]
psnr = [37.18, 38.34, 39.39, 39.19]
ours, = plt.plot(bpp, psnr, "k-o", linewidth=LineWidth, label='Proposed')

# H.264 very fast setting.
bpp = [0.01431713, 0.019554398, 0.025353009, 0.041417824]
psnr = [37.25, 37.47, 37.74, 38.25]
h264, = plt.plot(bpp, psnr, "m--s", linewidth=LineWidth, label='H.264')

bpp = [0.012702546, 0.019392361, 0.025052083, 0.042853009]
psnr = [37.95, 38.23, 38.47, 38.8]
h265, = plt.plot(bpp, psnr, "r--v", linewidth=LineWidth, label='H.265')


plt.legend(handles=[h264, h265, ours], loc=4)
plt.grid()
plt.xlabel('Bpp')
plt.ylabel('PSNR(dB)')
# plt.title('')
plt.savefig('ClassB_PSNR.eps', format='eps', dpi=300, bbox_inches='tight')
```
